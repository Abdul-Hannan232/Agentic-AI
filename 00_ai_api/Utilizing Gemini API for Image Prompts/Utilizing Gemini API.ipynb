# 3.

Let's configure the Gemini API, Gemini 1.5 Flash. For now, we are configuring it at a very basic level. In future we will also learn how to give set Temperature and Tokens.

 Let's your first prompt using the Gemini 1.5 Flash LLM model and give it some prompt.

```python
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content("Give me a short guide to prompt images as input to gemini flash 1.5 api on Colab using google-generativeai 0.7.2 sdk")
print(response.text)
```

# 6.

Write Prompt and store in a variable.

```python
prompt = """The image I am providing to you contains information of a product I am willing to sale on ecommerce platforms.
Extract the complete information from the image. Return output in json format:
{description: description, features: [feature1, feature2, feature3, etc]}"""
```

# 7.
